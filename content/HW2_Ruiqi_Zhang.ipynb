#### Solution

Use the expansion function to expand the expression and keep decimal parts. Then use sliding_windows to get the n-digits number list we need to test. Finally, use judge_prime function to find the first n-digits number that is prime. The judge_prime function uses Sieve of Eratosthenes to find all primes that less than the sqrt(number), and then test whether this number is divisible by these primes. If this number is not divisible for all primes that less than sqrt(number), then this number is a prime.

```python
"""Import required package"""

import numpy as np
import sympy as sym
```

```python
def expansion(expression, length):
    """Expand expression and keep decimal parts

    Parameter
    ----------------------
    expression : math expression
        Required math expression
    length : int
        Required expand length
    
    Returns 
    ----------------------
    result : str
        Obtain expanded decimal part as string.
    """
    
    decimal = expression - int(expression)
    result = str(decimal.evalf(length+1))[2:-1]
    return result
```

```python
"""Test function expansion"""

expansion(sym.pi,100)
```

```python
def judge_prime(n):
    """Judge whether a number is prime
    
    Parameter
    ----------------------
    n : int
        number that need to be judged
    prime_list: list
        Contain all prime that less than sqrt(n)
    
    Returns 
    ----------------------
    True or False : Bool
        Whether this number is prime
    """
    
    judge_list = [True] * int(np.sqrt(n)+2)
    for i in range(2,len(judge_list)):
        if judge_list[i] == True:
            for j in range(i*i,len(judge_list),i):
                judge_list[j] = False
    prime_list = [number for number in range(2,len(judge_list)) if judge_list[number]]
    for prime in prime_list:
        if n % prime == 0:
            return False
    return True
```

```python
"""Test judge_prime function"""

judge_prime(1415926535897)
```

```python
def sliding_windows(string, n):
    """Generate required number list from input string
    
    Parameter
    ----------------------
    string : str
        a long number string (generated by expansion function)
    n : int
        Required number digits
    
    Returns 
    ----------------------
    result : list
        Generated n-digits number list using sliding windows
    """
    
    if len(string) < n:
        return None
    result = []
    for i in range(len(string)-n+1):
        result.append(int(string[i:i+n]))
    return result
```

```python
"""Test sliding_windows"""

sliding_windows('18364827476',4)
```

```python
def first_n_digit_prime(expression, n):
    """Find the first n-digits prime in expansion of expression
    
    Parameter
    ----------------------
    expression : math expression
        Required math expression
    n : int
        Required number digits
    requirement: int
        Criteria for judging whether the loop should continues
    
    Returns 
    ----------------------
    result : int 
        The first n-digits prime in expansion of expression
    """
    
    requirement = 1
    while requirement:
        times = 1
        string = expansion(expression, 100*n*times)[(times-1)*100*n-n*(times > 1):]
        number_list = sliding_windows(string, n)
        for number in number_list:
            if judge_prime(number) == True:
                requirement = 0
                result = number
                break
        times += 1
    return result
```

```python
"""Test function first_n_digit_prime"""

first_n_digit_prime(sym.exp(1),10)
```

```python
"""Obtain the required number"""

first_n_digit_prime(sym.pi*17, 10)
```
