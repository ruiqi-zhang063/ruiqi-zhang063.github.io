#### Description

This homework uses table 15.


Firstly, it uses openppyxl to obtain the indent in excel, then based on the indent to divide the hierarchical relationship in index.


The dashboard has three figures, one drop-down box and one time slider. The drop-down can choose male data, female data or total data,while the time slider can choose year. The first bar figure can show the doctorate recipents number by broad field this year. When we move the cursor to the histogram of a certain field, the second bar figure will show the subdivisions of doctorate recipents in this field, and the third line chart will show this field doctorate recipents every year.

We can see that the column "Sex and major field of study" has different levels. In original excel table, this level can be divided by indent. Therefore, we want to increase a column "indent" to indicate the hierarchical relationship.

#### The indent of data decides the index level, using openpyxl to obtain indent of the first column

```python
worksheet = load_workbook('data_tables/sed17-sr-tab015.xlsx')['Table 15']
indent = []
for cell in worksheet['A'][4:]:
    indent.append(cell.alignment.indent)
```

#### Insert indent to dataframe and change the columns name to string

```python
column_names = [index for index in data.columns]
column_names.insert(1, 'indent')
column_names.insert(1, 'Broad Field')
data_indent = data.reindex(columns = column_names)
data_indent['indent'] = indent
column_names = {year:str(int(year)) for year in column_names if type(year) == float}
column_names
data_indent = data_indent.rename(columns = column_names)
```

#### Divide data based on gender

```python
data_indent_total = data_indent[:45].copy()
data_indent_male = data_indent[45:90].copy()
data_indent_female = data_indent[90:].copy()
```

#### Insert Broad Field to dataframe

```python
temp = 'Na'
for i in data_indent_total.index:
    if data_indent_total['indent'][i] == 1:
        temp = data_indent_total['Sex and major field of study'][i]
    elif data_indent_total['indent'][i] == 2:
        data_indent_total['Broad Field'][i] = temp

for i in data_indent_male.index:
    if data_indent_male['indent'][i] == 2:
        temp = data_indent_male['Sex and major field of study'][i]
    elif data_indent_male['indent'][i] == 3:
        data_indent_male['Broad Field'][i] = temp
        
for i in data_indent_female.index:
    if data_indent_female['indent'][i] == 2:
        temp = data_indent_female['Sex and major field of study'][i]
    elif data_indent_female['indent'][i] == 3:
        data_indent_female['Broad Field'][i] = temp
```

#### Pandas Analysis

```python
analysis_data = data_indent_total[data_indent_total['indent'] == 1].copy()
analysis_data.sort_values(by = '2008',ascending = False)
```

```python
analysis_data.sort_values(by = '2017',ascending = False)
```

We can see that compared to 2008, there are less doctorate recipients in the field "Education" in 2017.

```python
fields = list(analysis_data['Sex and major field of study'])
Rate_dict = {}
for field in fields:
    male_number = float(data_indent_male['2017'][data_indent_male['Sex and major field of study'] == field])
    female_number = float(data_indent_female['2017'][data_indent_female['Sex and major field of study'] == field])
    Rate_dict[field] = male_number / female_number
pd.DataFrame({'Field':Rate_dict.keys(),r'Rate (Male / Female)':Rate_dict.values()}).sort_values(by = r'Rate (Male / Female)',ascending = False)
```

We can see that in these fields, engineering has the highest rate, while education has the lowest rate.

#### Run the dashboard

```python
external_stylesheets = ['https://codepen.io/chriddyp/pen/bWLwgP.css']

app = dash.Dash(__name__, external_stylesheets=external_stylesheets)

available_indicators = ['Total','Male','Female']

app.layout = html.Div([
    
    html.H1(
        children='Doctorate Recipients by Sex and Major Field',
        style={
            'textAlign': 'center'
        }
    ),
    
    html.Div([
        html.Div([
            dcc.Dropdown(
                id='Gender',
                options=[{'label': i, 'value': i} for i in available_indicators],
                value='Total'
            )
        ],
        style={'width': '49%', 'display': 'inline-block'}),
    ], style={
        'padding': '10px 5px'
    }),

    html.Div([
        'Doctorate Recipients by Broad Field',
        dcc.Graph(
            id='Broad-Field-Bar',
            hoverData={'points': [{'customdata': 'Engineering'}]}
        )
    ], style={'width': '49%', 'display': 'inline-block', 'padding': '0 20'}),
    
    html.Div([
        'Doctorate Recipients in Specific Field',
        dcc.Graph(id='Specific-Field-Bar'),
    ], style={'display': 'inline-block', 'width': '49%'}),
    
    html.Div(dcc.Slider(
        id='Year-Slider',
        min=2008,
        max=2017,
        value=2017,
        marks={str(year): str(year) for year in range(2008, 2018)},
        step=None
    ), style={'width': '49%', 'padding': '0px 20px 20px 20px'}),
    
    html.Div([
        'Doctorate Recipients Every Year in Broad Field',
        dcc.Graph(id='Time-Series'),
    ]),
])

@app.callback(
    dash.dependencies.Output('Broad-Field-Bar', 'figure'),
    [dash.dependencies.Input('Gender', 'value'),
     dash.dependencies.Input('Year-Slider', 'value')
    ])
def update_graph(Gender, Year):
    if Gender == 'Total':
        dff = data_indent_total[data_indent_total['indent'] == 1]
    elif Gender == 'Male':
        dff = data_indent_male[data_indent_male['indent'] == 2]
    else:
        dff = data_indent_female[data_indent_female['indent'] == 2]

    fig = px.bar(dff, x = 'Sex and major field of study', y = str(Year))

    fig.update_traces(customdata = dff['Sex and major field of study'])
    
    fig.update_xaxes(title = 'Broad Field')
    
    fig.update_yaxes(title='Doctorate Recipients Number')

    fig.update_layout(margin={'l': 40, 'b': 40, 't': 10, 'r': 0}, hovermode='closest')

    return fig

@app.callback(
    dash.dependencies.Output('Specific-Field-Bar', 'figure'),
    [dash.dependencies.Input('Broad-Field-Bar', 'hoverData'),
     dash.dependencies.Input('Gender', 'value'),
     dash.dependencies.Input('Year-Slider', 'value')
    ])
def update_Specific_Field_Bar(hoverData, Gender, Year):
    if Gender == 'Total':
        dff = data_indent_total[data_indent_total['Broad Field'] == hoverData['points'][0]['customdata']]
    elif Gender == 'Male':
        dff = data_indent_male[data_indent_male['Broad Field'] == hoverData['points'][0]['customdata']]
    else:
        dff = data_indent_female[data_indent_female['Broad Field'] == hoverData['points'][0]['customdata']]
    
    fig = px.bar(dff, x = 'Sex and major field of study', y = str(Year))
    
    fig.update_xaxes(title = 'Specific Field in ' + hoverData['points'][0]['customdata']) 
    
    fig.update_yaxes(title='Doctorate Recipients Number')

    return fig

@app.callback(
    dash.dependencies.Output('Time-Series', 'figure'),
    [dash.dependencies.Input('Broad-Field-Bar', 'hoverData'),
     dash.dependencies.Input('Gender', 'value'),
    ])

def create_time_series(hoverData, Gender):
    
    if Gender == 'Total':
        dff = data_indent_total[data_indent_total['indent'] == 1]
    elif Gender == 'Male':
        dff = data_indent_male[data_indent_male['indent'] == 2]
    else:
        dff = data_indent_female[data_indent_female['indent'] == 2]
        
    dff = dff[dff['Sex and major field of study'] == hoverData['points'][0]['customdata']]
    
    x = range(2008,2018)
    
    y = list(dff.iloc[0])[3:-1]

    fig = px.scatter(x = x, y = y)

    fig.update_traces(mode='lines+markers')

    fig.update_xaxes(showgrid=False)
    
    fig.update_xaxes(title = 'Year')
    
    fig.update_yaxes(title='Doctorate Recipients Number')
    
    fig.add_annotation(x=0, y=0.85, xanchor='left', yanchor='bottom',
                       xref='paper', yref='paper', showarrow=False, align='left', 
                       text = 'Doctorate Recipients Number in ' + hoverData['points'][0]['customdata'])

    return fig

if __name__ == '__main__':
    app.run_server()
```
